```python
import json
import pandas
```

```python
# with open("params.json", "r") as file:
#     data = json.load(file)

filepath = r"Z:\Dropbox\CurrentClasses\graduate_project\AOI\UpdatedGraphAOI.tsv"
df = pandas.read_csv(filepath, index_col=False, sep="\t")
df
```

```python
dict1 = {}
columns = df.columns.values
print(columns)

for index, col in df.iterrows():
    graph_type = col["GraphType"]
    dict2 = {}
    # Every 4 reading: top left, bottom left, top right, bottom right
    max_len = len(col)
    for i in range(1, max_len, 4):
        # Dictionary key will be the label of the AOI
        aoi_label = columns[i][:-2]
        # Top Left (x, y) coordinates
        top_left = [a.strip() for a in col[i].split(",")]
        if len(top_left) > 2:
            raise ValueError("There should only be two values")
        top_left = ",".join(top_left)
        # Bottom left (x, y) coordinates
        bot_left = [a.strip() for a in col[i + 1].split(",")]
        if len(bot_left) > 2:
            raise ValueError("There should only be two values")
        bot_left = ",".join(bot_left)
        # Top Right (x, y) coordinates
        top_right = [a.strip() for a in col[i + 2].split(",")]
        if len(top_right) > 2:
            raise ValueError("There should only be two values")
        top_right = ",".join(top_right)
        # Bot Right (x, y) coordinates
        bot_right = [a.strip() for a in col[i + 3].split(",")]
        if len(bot_right) > 2:
            raise ValueError("There should only be two values")
        bot_right = ",".join(bot_right)
        
        dict2[aoi_label] = [top_left, bot_left, top_right, bot_right]
    dict1[graph_type] = dict2
    
print(json.dumps(dict1, indent=2))

# with open("blah2.json", "w") as file:
#     json.dump(dict1, file, indent=2)
```
